## Touring the Container

### Managing Virtual Environments

The aima-docker container uses `conda` to manage multiple python virtual environments. All machine learning frameworks and their dependencies are installed in separate virtual environments. Currently aima-docker provides the following environments:

 - **tensorflow:** Contains latest versions of tensorflow and keras
 - **pytorch:** Contains latest pytorch
 - **cntk:** Contains latest versions of cntk and keras
 
The base/root environment contains all packages required to run aima-python. Package dependencies can be managed on the `Conda` tab of Jupyter Notebook home. To install new packages in an environment, simply go to the `Conda` tab and click on the environment. You should then be able to edit the environment packages in the package manegement window highlighted below.

![Package Management Screen](images/env_management.png)

While the package management window allows you to easily create environments and manage their packages, it only lists the packages that are available on official conda channels. In many cases, you will need to install packages from PyPi, other conda channels, or even build them from source. To enable this, Jupyter can be used to provide terminal access to the container. You can start a terminal just like you start a new notebook.

![How to start a Terminal](images/terminal_start.png)

Clicking on `Terminal` in `new` menu gives us the bash we all know and love.

![Terminal Screenshot](images/terminal.png)

However, you must keep in mind that the additional packages installed on your container will not persist across Jupyter sessions and will be deleted whenever you restart the container. If you want a package to be permanently added to aima-docker, raise a Github issue detailing why the package is needed. 

### Persistent Storage for Custom Models

We know you might want to create your own Jupyter Notebooks, execute your own code, or import datasets into the container. Good News! It is very simple to do so. Whenever you are working with aima-docer, the `code` folder in the container synchronizes with the folder at `~/aima-docker` on your system. This means that everything you import in this folder will be available on the aima-docker container and persist even if you restart it.

Coupled with your favorite text editor and the Jupyter bash, you can create and execute any scripts you like on the aima-docker container. We also recommend that you keep any data your models require in this folder. This is because everything in the folder's subtree is persistent, and doing so will save your scripts from downloading data everytime the container is restarted.

### Terminating the Container 

aima-docker container runs as a daemon by default. Don't worry though, terminating the container is easy. Go to Jupyter home and click the `Quit` button. That's it, you're done.

![Closing aima-docker](images/quit_jupyter.png)

## FAQs

### Jupyter is running with authentication disabled, isn't this a security flaw?

While Jupyter Notebooks running without authentication is a security flaw, this is not so for the aima-docker container. Authentication is required since anyone with access to Jupyter on your network can essentially execute any code on your machine. However, since we are running aima-docker inside a non-privilaged container with a non-root user, your machine is isolated from whatever code is being executed on Jupyter. Therefore, anyone with access to aima-docker's Jupyer will not be able to execute arbitrary codes on your machine.

### Contributors run Jupyter as a root user, won't it compromize security?

For contributors, token authentication is enabled. Therefore, only they can access their Jupyter notebooks from their browsers.


## My changes don't persist across different sessions

Docker containers are meant to be ephemeral and stateless i.e. they are meant to be  created and destroyed often without any changes in state. To give some persistent storage, aima-docker provides for a `code` folder which is bound to your local system. Read **Persistent Storage for Custom Models** section above for more information.

If you're a docker vituoso, you can simply remove the *--rm* flag from the docker run command. This will not delete your container whenever it closes. You can the use

    docker start *container name*

to restart the same container. Needless to say, *all* your changes will be retained.
